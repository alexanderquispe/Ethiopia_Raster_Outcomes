```python
from types_data.utils import RasterIOInd
import geopandas as gpd, warnings
warnings.filterwarnings('ignore')
```

```python
osm_raster_path = "../Data/Raster/osm/eth_osm_dst_road_100m_2016.tif"
virrs_raster_path = '../Data/Raster/viirs_100m/eth_viirs_100m_2016.tif'
dist_data = gpd.read_file("../Data/gjson/adm_3.geojson").sort_values('area')
```

```python
osm = RasterIOInd(
    osm_raster_path,
    dist_data,
    "distance_osm",
    "../output/osm/distance_osm.csv",
    only_metric=True,
)
osm.get_result()
print(osm.total_time)
osm.result_df.head()
```

```python
virrs = RasterIOInd(
    virrs_raster_path,
    dist_data,
    "night_time",
    "../output/virrs/night_time.csv",
    only_metric=True,
)
virrs.get_result()
virrs.total_time
virrs.result_df
```
