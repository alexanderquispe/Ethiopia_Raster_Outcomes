```python
from types_data.utils import RasterIOInd
import geopandas as gpd, warnings
warnings.filterwarnings('ignore')



decision = "../Data/Raster/ETH_Maternal_and_child_socioeconomic/ETH_DECISION_MEAN.tif"
hwealth = "../Data/Raster/ETH_Maternal_and_child_socioeconomic/ETH_HWEALTH_MEAN.tif"
education = "../Data/Raster/ETH_Maternal_and_child_socioeconomic/ETH_MEDUCATION_MEAN.tif"
dist_data = gpd.read_file("../Data/gjson/adm_3.geojson").sort_values('area')
```
